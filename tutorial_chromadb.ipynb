```python
import chromadb

client = chromadb.Client()
collection = client.create_collection(name="new_collection")
```

```python
collection.add(
    documents=[
        "this is a test document",
        "this is a statemtn about delhi"
    ],
    ids = ["doc1", "doc2"]
)
```

```python
all = collection.get()
all
```

```python
docu1 = collection.get(ids=["doc1"])
docu1
```

```python
results = collection.query(
    query_texts=['query is about air pollution'],
    n_results=2
)

results
```

```python
collection.delete(ids=all['ids'])
collection.get()
```
